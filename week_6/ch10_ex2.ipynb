```python
import matplotlib.pyplot as plt
```

```python
flu_case = [13, 14, 9, 16, 10, 18, 22, 19, 16, 22,
             24, 48, 34, 25, 17, 29, 33, 35, 28, 43,
             59, 44, 55, 63, 61, 48, 68, 61, 70, 76,
             78, 74, 87, 101, 120, 128, 105, 109, 120, 124,
             111, 128, 120, 133, 134, 139, 127, 130, 141, 147,
             439, 236, 218, 209, 213, 244, 329, 197, 351, 325]

days = range(len(flu_case))
```

```python
# 7 days moving average
# # first week-average
# sum(flu_case[0:7])/7
# # moving by 1 day
# sum(flu_case[0+1:7+1])/7
# # ...
# sum(flu_case[len(flu_case)-7: len(flu_case)])/7

week_average = []
for i in range(6,len(flu_case)):
    week_average.append( sum(flu_case[i-7 : i])/7 )
```

```python
plt.plot(days, flu_case, ':o', markersize=2, label = "daily")
plt.plot(days[6:len(days)], week_average, '-', label = "7-day average")
plt.xlabel('day')
plt.ylabel('covid cases')
# fig.savefig("delete_me.tiff", dpi = 1200)
```
